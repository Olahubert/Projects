```python
from sklearn.preprocessing import LabelEncoder
```

```python
#creating an object from the Label Encoding class for Gender
Le = LabelEncoder()
Loan["new gender"] = Le.fit_transform(Loan["Gender"])
```

```python
Loan["new_married"] = Le.fit_transform(Loan["Married"])
```

```python
Loan["new_education"] = Le.fit_transform(Loan["Education"])
```

```python
Loan["new_self_employed"] = Le.fit_transform(Loan["Self_Employed"])
```

```python
Loan["new_property_area"] = Le.fit_transform(Loan["Property_Area"])
```

```python
Loan.dropna(inplace=True)
```

```python
#Extract the value of X
X = Loan[['Dependents', 'ApplicantIncome', 'CoapplicantIncome',
      'LoanAmount', 'Loan_Amount_Term', 'Credit_History',
       'new gender', 'new_married','new_education','new_self_employed','new_property_area']].values
```

```python
#Extract the value of Y
Y = Loan['Loan_Status'].values
```

```python
#Split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Build a logistic Regression by training
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#Make pridiction
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print("Accuracy:",accuracy_score(Y_test,Y_pred))
print("Confusion Matrix: Report:\n", confusion_matrix(Y_test,Y_pred))
print("Classification Report:\n", classification_report(Y_test,Y_pred))
```
